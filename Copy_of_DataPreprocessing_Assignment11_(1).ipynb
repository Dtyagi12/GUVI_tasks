**Step 1: Importing the libraries**

**Step 3: Handling the missing data**

```python
data['Age'].fillna(data.Age.mean(), inplace=True)
data
```

```python
data['Salary'].fillna(data.Salary.mean(), inplace=True)
data
```

**Step 4: Encoding categorical data**

```python
from sklearn import preprocessing
label_encode=preprocessing.LabelEncoder()
#df["Country"]=label_encode.fit_transform(df["Country"])
data["Purchased"]=label_encode.fit_transform(data["Purchased"])
```

```python
data
```

**Step 5: Creating a dummy variable**

```python
data=pd.get_dummies(data, prefix='Country', columns=['Country'])
data
```

```python
x=data[["Country_France","Country_Germany","Country_Spain","Age","Salary"]].values
y=data["Purchased"].values
```
